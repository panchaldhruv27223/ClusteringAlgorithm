```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=23)
X.shape, y.shape
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
X_train.shape,X_test.shape
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow method
## manualy ## t selcect the k value
```

```python
wcss = []

for i in range(1,11):
    kmean = KMeans(n_clusters=i,init="k-means++")
    kmean.fit(X_train)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("number of cluster")
plt.xticks(range(1,11))
plt.ylabel("wcss")
plt.show()
```

```python
kmean = KMeans(n_clusters=3,init="k-means++")
```

```python
y_label = kmean.fit_predict(X_train)
```

```python
y_label
```

```python
plt.scatter(X_train[:,0],X_train[:,1])
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_label)
```

```python
y_test_label = kmean.predict(X_test)
```

```python
y_test_label
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test_label)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette__score = []

for i in range(2,11):
    kmean = KMeans(n_clusters=i,init="k-means++")
    kmean.fit(X_train)
    score = silhouette_score(X_train,kmean.labels_)
    silhouette__score.append(score)
```

```python
silhouette__score
```

```python
plt.plot(range(2,11),silhouette__score)
plt.xlabel("number of cluster")
plt.xticks(range(2,11))
plt.ylabel("score")
plt.show()
```
